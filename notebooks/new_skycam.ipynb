```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

import astropy.visualization as viz
```

```python
%load_ext autoreload
%autoreload 2
%matplotlib widget
```

```python
hdul = fits.open("../data/2022_10_27__19_29_35.fits")
red = hdul[0].data[0, :, :]
green = hdul[0].data[1, :, :]
blue = hdul[0].data[2, :, :]
```

```python
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(1, 1, 1)
im, norm = viz.imshow_norm(green, ax, origin='lower', interval=viz.PercentileInterval(98), stretch=viz.SqrtStretch())
fig.colorbar(im)
plt.show()
```

```python
rgb = viz.make_lupton_rgb(1.0*red, 0.95*green, 1.0*blue, minimum=1950, stretch=50, Q=20)
plt.imshow(rgb)
plt.show()
```

```python
a = np.transpose(hdul[0].data, axes=(1, 2, 0)) - 1950
a = a / a.max()
a = a ** 0.3
plt.imshow(a)
plt.show()
```
